```python
import pandas as pd
from pathlib import Path

ROOT = Path("..").resolve()
consensus = pd.read_csv(ROOT / "results" / "consensus_biomarkers.csv")
print(consensus.head(10))
```

```python
# bioloical annotation of top genes

known_ifn_genes = {
    "ISG15":    "Ubiquitin-like IFN-stimulated gene; hallmark of IFN signaling; proposed lupus biomarker",
    "IFIT3":    "Interferon-induced; blocks viral replication",
    "LY6E":     "IFN-induced; restricts viral entry",
    "ISG20":    "IFN-stimulated exonuclease; degrades viral RNA",
    "IFIT1":    "Interferon-induced; inhibits viral translation",
    "IFIT2":    "Interferon-induced; inhibits viral translation",
    "IRF7":     "Interferon Regulatory Factor 7; master transcription factor driving IFN production",
    "OAS1":     "2'-5'-oligoadenylate synthetase; activates RNase L to degrade viral RNA",
    "CXCL10":   "IFN-induced chemokine; recruits immune cells; elevated in lupus patients",
    "APOBEC3A": "IFN-induced cytidine deaminase; restricts retroviruses",
}

consensus["known_role"] = consensus["gene_symbol"].map(known_ifn_genes)
consensus["validated"] = consensus["known_role"].notna()

print(f"Validated by literature: {consensus['validated'].sum()} / {len(consensus)} top genes")
print(consensus[["gene_symbol", "n_cell_types", "validated"]])
```

```python
# save annotated table

consensus.to_csv(ROOT / "results" / "validated_biomarkers.csv", index=False)
print(consensus[consensus["validated"]][["gene_symbol", "n_cell_types", "known_role"]])
```
